```python
import ast
```

```python
#this function helps in converting the above list of dictionaries (i.e.genres) into required format
# ex:- ['Action','Adventure','Fantasy','Science Fiction']
def convert(text):
    L = []
    #here the above text is a string of list. ast.literal_eval converts it into a simple list
    for i in ast.literal_eval(text):
        L.append(i['name']) 
    return L 
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
movies.head()
```

```python
#This function is to fetch the name of top 3 actors (cast) of the movie
def convert3(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
        counter+=1
    return L 
```

```python
movies['cast'] = movies['cast'].apply(convert)
movies.head()
```

```python
#This function is to fetch the name of director of the movie
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L 
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
movies['overview'][0]
```

```python
#so we can see here that overview is a string and to convert it into list we are writing following lines of code
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
#To remove the spaces between the words for ex. Sam Worthingtom-> SamWorthington this is to avoid confusion and mis-predictions
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
```

```python
movies.head()
```

```python
#to merge overview, genre, keywords, cast and crew of the movie into a single column known as "tags"
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
new_df = movies.drop(columns=['overview','genres','keywords','cast','crew'])
```

```python
new_df
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: " ".join(x))
new_df.head()
```

```python
new_df['tags'][0]
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.head()
```

```python
#To convert the above text into vectors we are using CountVectorizer from sklearn
#Here we are using bag of words technique to convert the above text into vectors wherein we merge tags of all the movies and from this we pick up 5000 most frequently occurring words.Then we count the frequency of these words in tags of each movie. So, the dimentionality is 5000*5000 i.e. 5000 movies in 5000 dimensional space
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english') #here we are not considering english stop words like (and, are, or, etc) in those 5000 words
```

```python
vector = cv.fit_transform(new_df['tags']).toarray()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
#This is the function to perform stemming which converts 'loving'->'love' & 'loved'->'love' & 'love'->'love'
def stem(text):
    y=[]

    for i in text.split():
        y.append(ps.stem(i))

    return " ".join(y)
```

```python
ps.stem('loved')
```

```python
ps.stem('loving')
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(vector)
```

```python
similarity=cosine_similarity(vector)
```

```python
similarity
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list=sorted(list(enumerate(distances)),reverse=True,key = lambda x: x[1])[1:6]
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Gandhi')
```
